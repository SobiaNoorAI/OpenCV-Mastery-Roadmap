```python
import cv2 as cv
```

Images I/O

```python
# load image in 2D array
earth=cv.imread('C:/Users/QC/Downloads/Earth Day 2026.png')
```

```python
# Display image
cv.imshow('Earth Day 2026', earth)
#pause image for 5 seconds
cv.waitKey(5000)
cv.destroyWindow('Earth Day 2026')
```

```python
#save image
new=cv.imwrite('Earth.jpg', earth )
print('Image saved successfully')
cv.imshow('Earth', new)
cv.waitKey(5000)
cv.destroyWindow('Earth')
```
