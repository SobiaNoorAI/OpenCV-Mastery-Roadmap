Drwaing 

```python
#load image
jerry=cv.imread('images/jerry.png')

cv.imshow('Jerry', jerry)
cv.waitKey(5000)
cv.destroyWindow('Jerry')
```

```python
#draw line on image copy image to avoid modifying original image
line_jerry=jerry.copy()
#cv.line(image, start_point, end_point, color, thickness, lineType, shift)
cv.line(line_jerry, (250,100), (600,100), (0,0,250), 4, 4)
cv.imshow('Jerry with line', line_jerry)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
img=cv.imread('Earth.jpg')
cv.imshow('Practice', img)
cv.waitKey(0)
cv.destroyAllWindows()

line_img= cv.line(img,(0,0),(1000, 1000),(0,0,255),10, 1)
cv.line(img, (1000,0), (0,1000), (0,255,0), 10, 16)
cv.imshow('line practice', line_img)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#draw rectangle on image copy
rect_jerry=jerry.copy()
#cv.rectangle(image, start_point, end_point, color, thickness, lineType, shift)
cv.rectangle(rect_jerry, (110,30), (550,400), (0,0,0), 4)
cv.imshow('Jerry with rectangle', rect_jerry)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#draw more rectangles on image copy
rect_jerry=jerry.copy()
#cv.rectangle(image, start_point, end_point, color, thickness, lineType, shift)
cv.rectangle(rect_jerry, (110,30), (550,400), (0,0,0), 4)
cv.rectangle(rect_jerry, (90, 500), (250, 600), (255,0,0),10)
cv.rectangle(rect_jerry, (500, 520), (670, 630), (0,0,255), 20)
cv.rectangle(rect_jerry, (440, 250), (470, 270), (200,220,230),-1)
cv.imshow('Jerry with rectangle', rect_jerry)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#draw circle on image copy
circle_jerry=jerry.copy()
#cv.circle(image, center_coordinates, radius, color, thickness, lineType, shift)
cv.circle(circle_jerry, (355,180), 250, (0,0,0), 1, 1 )
cv.circle(circle_jerry, (200, 150), 30, (255,0,0), -1)
cv.imshow('Jerry with circle', circle_jerry)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import numpy as np

#draw polylines on image copy
polylines_jerry=jerry.copy()
#cv.polylines(image, points, isClosed, color, thickness, lineType, shift)
points = np.array([[100,100], [200,100], [200,200],[150,150], [150,200], [100,200]], dtype=np.int32)
points2 = np.array([[300, 300], [400, 300], [300, 400],[400,400]], np.int32)
cv.polylines(polylines_jerry, [points], True, (255,0,0), 3)
cv.polylines(polylines_jerry, [points2], True, (0,0,244), 10)
cv.imshow('jerry with Polylines',polylines_jerry)
cv.waitKey(0)       
cv.destroyAllWindows()
```

```python

#draw annotations on image copy
annotation_jerry=jerry.copy()
#cv.arrowedLine(image, start_point, end_point, color, thickness, lineType, shift, tipLength)
cv.arrowedLine(annotation_jerry, (460,280), (600,250), (0,0,250), 4, cv.LINE_4, 0, 0.1)
cv.arrowedLine(annotation_jerry, (400,150), (200, 100), (0,255,0), 10, 16,0, 0.1)
cv.imshow('Jerry with annotation', annotation_jerry)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#write text on image copy
text_jerry=jerry.copy()
#cv.putText(image, text, org, fontFace, fontScale, color, thickness, lineType, bottomLeftOrigin)
cv.putText(text_jerry, 'Happy Jerry', (100,25), cv.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2, cv.LINE_8, True)
cv.imshow('Jerry with text', text_jerry)
cv.waitKey(0)
cv.destroyAllWindows()
```
