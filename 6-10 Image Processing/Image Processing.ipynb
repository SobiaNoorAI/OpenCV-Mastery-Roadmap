```python
import cv2 as cv
import numpy as np
```

Image resize

```python
# Load the image
image = cv.imread('H:/OpenCV-Mastery-Roadmap/1-5 The Basics/images/Earth.jpg')
#check current size
image.shape
```

```python
#Resize the image
width, height = 500, 500
resized_image = cv.resize(image, (width, height))

#show the resized image
cv.imshow('Resized Image', resized_image)
cv.imshow('Original Image', image)
cv.waitKey(0)
cv.destroyAllWindows()
```

Image scaling 

```python
#load an image
img=cv.imread('H:\OpenCV-Mastery-Roadmap\1-5 The Basics\images\atm.png')
img = cv.imread(r'H:\OpenCV-Mastery-Roadmap\1-5 The Basics\images\atm.png') # using raw string to avoid escape character issues
if img is None:
    raise FileNotFoundError(r'H:\OpenCV-Mastery-Roadmap\1-5 The Basics\images\atm.png')

#show original image
cv.imshow('ATM', img)

#resize image
#cv.resize(src, dsize, fx, fy, interpolation)
img_linear=cv.resize(img, None, fx=3, fy=3, interpolation=cv.INTER_LINEAR)
img_area=cv.resize(img, None, fx=3, fy=3, interpolation=cv.INTER_AREA)
cv.imshow('Linear Interpolation', img_linear)
cv.imshow('Area Interpolation', img_area)
cv.waitKey(0)
cv.destroyAllWindows()
```

Image Crop

```python
#image cropping
cropped_image = image[140:270, 300:800]
cv.imshow('Cropped Image', cropped_image)
cv.waitKey(0)
cv.destroyAllWindows()
```

Image Translation

```python
#translation
#matrix for translation
matrix = np.float32([[1,0,100], [0,1,50]])
#cv.warpAffine(src, M, dsize)
img_translation = cv.warpAffine(img, matrix, (img.shape[1]+100, img.shape[0]+50)) 

cv.imshow('Translation', img_translation)
cv.imshow('Original Image', img)
cv.waitKey(0)
cv.destroyAllWindows()
```

Image Rotation

```python
#rotation
#matrix for rotation
#centter = (width//2, height//2)
center = (img.shape[1]//2, img.shape[0]//2)
#cv.getRotationMatrix2D(center, angle, scale)
matrix = cv.getRotationMatrix2D(center, 45, 1) # rotate 45 degrees with scale of 1

#
img_rotation = cv.warpAffine(img, matrix, (img.shape[1], img.shape[0]))
cv.imshow('Rotation', img_rotation)
cv.imshow('Original Image', img)
cv.waitKey(0)
cv.destroyAllWindows()
```

Perspective Transform

```python
#perspective transform
#points in the original image
pts1 = np.float32([[39,38], [93, 36],  [39,55], [94,50]])
#points in the output image
# height, width of output image is 300x300
pts2 = np.float32([[0,0], [300,0], [0,300], [300,300]])
    
# get the perspective transform matrix    
matrix=cv.getPerspectiveTransform(pts1, pts2) 
img_perspective = cv.warpPerspective(img, matrix, (300,300)) # apply perspective transform
cv.imshow('Perspective Transform', img_perspective)
cv.imshow('Original Image', img)
cv.waitKey(0)
cv.destroyAllWindows()
```
