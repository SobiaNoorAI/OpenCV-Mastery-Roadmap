```python
import cv2 as cv
import numpy as np
```

Image resize

```python
# Load the image
image = cv.imread('H:/OpenCV-Mastery-Roadmap/images/Earth.jpg')
#check current size
image.shape
```

```python
#Resize the image
width, height = 500, 500
resized_image = cv.resize(image, (width, height))

#show the resized image
cv.imshow('Resized Image', resized_image)
cv.imshow('Original Image', image)
cv.waitKey(0)
cv.destroyAllWindows()
```

Image scaling 

```python
#load an image
img=cv.imread('H:\OpenCV-Mastery-Roadmap\images\atm.png')
img = cv.imread(r'H:\OpenCV-Mastery-Roadmap\images\atm.png') # using raw string to avoid escape character issues
if img is None:
    raise FileNotFoundError(r'H:\OpenCV-Mastery-Roadmap\images\atm.png')

#show original image
cv.imshow('ATM', img)

#resize image
#cv.resize(src, dsize, fx, fy, interpolation)
img_linear=cv.resize(img, None, fx=3, fy=3, interpolation=cv.INTER_LINEAR)
img_area=cv.resize(img, None, fx=3, fy=3, interpolation=cv.INTER_AREA)
cv.imshow('Linear Interpolation', img_linear)
cv.imshow('Area Interpolation', img_area)
cv.waitKey(0)
cv.destroyAllWindows()
```

Image Crop

```python
#image cropping
#cropping is done by slicing the image array
#cropped_image = image[y1:y2, x1:x2]
cropped_image = image[140:270, 300:800]
cv.imshow('Cropped Image', cropped_image)
cv.imshow('Original Image', image)
cv.waitKey(0)
cv.destroyAllWindows()
```

Image Translation

```python
#translation
#matrix for translation
matrix = np.float32([[1,0,100], [0,1,50]])
#cv.warpAffine(src, M, dsize)
img_translation = cv.warpAffine(img, matrix, (img.shape[1]+100, img.shape[0]+50)) 

cv.imshow('Translation', img_translation)
cv.imshow('Original Image', img)
cv.waitKey(0)
cv.destroyAllWindows()
```

Image Rotation

```python
#rotation
#matrix for rotation
#centter = (width//2, height//2)
center = (img.shape[1]//2, img.shape[0]//2)
#cv.getRotationMatrix2D(center, angle, scale)
matrix = cv.getRotationMatrix2D(center, 45, 1) # rotate 45 degrees with scale of 1

#
img_rotation = cv.warpAffine(img, matrix, (img.shape[1], img.shape[0]))
cv.imshow('Rotation', img_rotation)
cv.imshow('Original Image', img)
cv.waitKey(0)
cv.destroyAllWindows()
```

Perspective Transform

```python
#perspective transform
#points in the original image
pts1 = np.float32([[39,38], [93, 36],  [39,55], [94,50]])
#points in the output image
# height, width of output image is 300x300
pts2 = np.float32([[0,0], [300,0], [0,300], [300,300]])
    
# get the perspective transform matrix    
matrix=cv.getPerspectiveTransform(pts1, pts2) 
img_perspective = cv.warpPerspective(img, matrix, (300,300)) # apply perspective transform
cv.imshow('Perspective Transform', img_perspective)
cv.imshow('Original Image', img)
cv.waitKey(0)
cv.destroyAllWindows()
```

Blurring

```python
#load image
img2=cv.imread('H:/OpenCV-Mastery-Roadmap/images/ironman.png')
#Gaussian Blur
#cv.GaussianBlur(src, ksize, sigmaX)

# Gaussian Blur
blurred_image = cv.GaussianBlur(img2, (5, 5), 10)
cv.imshow('Gaussian Blur', blurred_image)
cv.imshow('Original Image', img2)
cv.waitKey(0)
cv.destroyAllWindows()
```

Vertical Stretching 

```python
#stretch image vertically by gaussian blur
blurred_image = cv.GaussianBlur(img2, (1, 25), 10)
cv.imshow('VerticalGaussian Blur', blurred_image)
cv.imshow('Original Image', img2)
cv.waitKey(0)
cv.destroyAllWindows()
```

Horizontal Stretching

```python
#stretch image horizontally by gaussian blur
blurred_image = cv.GaussianBlur(img2, (25, 1), 25)
cv.imshow('HorizontalGaussian Blur', blurred_image)
cv.imshow('Original Image', img2)
cv.waitKey(0)
cv.destroyAllWindows()
```

Median Blur

```python
#median blur
#cv.medianBlur(src, ksize)
median_blurred = cv.medianBlur(img2, 5)
cv.imshow('Median Blur', median_blurred)
cv.imshow('Original Image', img2)
cv.waitKey(0)
cv.destroyAllWindows()
```

Thresholdoing

```python
#adaptive thresholding
#cv.adaptiveThreshold(src, maxValue, adaptiveMethod, thresholdType, blockSize, C)
#load image in grayscale for thresholding
img3=cv.imread('H:/OpenCV-Mastery-Roadmap/images/text.jpeg')
gray_img = cv.cvtColor((img3), cv.COLOR_BGR2GRAY)
adaptive_thresh = cv.adaptiveThreshold(gray_img, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 11, 2)
cv.imshow('Adaptive Thresholding', adaptive_thresh)
cv.imshow('Original Image', img3)
cv.waitKey(0)
cv.destroyAllWindows()
```

Edge Detection 

```python
#soble 
#cv.Sobel(src, ddepth, dx, dy, ksize)
gray = cv.cvtColor(img2, cv.COLOR_BGR2GRAY)
sobelx = cv.Sobel(gray, cv.CV_64F, 1, 0, ksize=5)
sobelx = cv.convertScaleAbs(sobelx)

cv.imshow('Sobel Edge Detection', sobelx)
cv.imshow('Original Image', img2)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#laplacian
#cv.Laplacian(src, ddepth, ksize)
laplacian = cv.Laplacian(gray, cv.CV_64F, ksize=5)
laplacian = cv.convertScaleAbs(laplacian)
cv.imshow('Laplacian Edge Detection', laplacian)
cv.imshow('Original Image', img2)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#canny edge detection
#cv.Canny(image, threshold1, threshold2)
edges = cv.Canny(gray, 100, 200)
cv.imshow('Canny Edge Detection', edges)
cv.imshow('Original Image', img2)
cv.waitKey(0)
cv.destroyAllWindows()
```

Contours

```python
#contours
#cv.findContours(image, mode, method)
#contours
#cv.findContours(image, mode, method)
img_jerry=cv.imread('H:/OpenCV-Mastery-Roadmap/images/jerry.png')
gray_for_contours = cv.cvtColor(img_jerry, cv.COLOR_BGR2GRAY)
_, gray_for_contours = cv.threshold(gray_for_contours, 127, 255, cv.THRESH_BINARY)
contours, hierarchy = cv.findContours(gray_for_contours, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)
cv.drawContours(img_jerry, contours, -1, (0,255,0), 3)
cv.imshow('Contours', img_jerry)
cv.waitKey(0)
cv.destroyAllWindows()
```
